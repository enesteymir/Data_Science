```python
seed(): sets the random seed, so that your results are the reproducible between simulations. As an argument, it takes an integer of your choosing
rand(): if you don't specify any arguments, it generates a random float between zero and one.
randint(x,y) : generates number between x and y but y will not be included
```

```python
# Import numpy as np
import numpy as np

# Set the seed
np.random.seed(123)

# Generate and print random float
print(np.random.rand())
```

```python
The following call generates the integer 4, 5, 6 or 7 randomly. 8 is not included.

import numpy as np
np.random.randint(4, 8)
```

```python
# Numpy is imported, seed is set

# Initialize random_walk
random_walk = [0]

for x in range(100) :   ##100 kere
    # Set step: last element in random_walk
    step = random_walk[-1]  ##son adımına eşit olsun

    # Roll the dice
    dice = np.random.randint(1,7)  ## za 1-6 arası atsın

    # Determine next step
    if dice <= 2:
        step = max(0,step - 1)   ## make sure that walk never be below 0
    elif dice <= 5:
        step = step + 1
    else:
        step = step + np.random.randint(1,7)

    # append next_step to random_walk
    random_walk.append(step)   ##random_walk a step ne ise onu ekliyoruz

# Print random_walk
print(random_walk)

# Import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

# Plot random_walk
plt.plot(random_walk)

# Show the plot
plt.show()
```

```python
# numpy and matplotlib imported, seed set.

# initialize and populate all_walks
all_walks = []
for i in range(10) :
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
            
        # Implement clumsiness
        if np.random.rand() <= 0.001 :
            step = 0
            
        random_walk.append(step)
    all_walks.append(random_walk)

# Convert all_walks to Numpy array: np_aw
np_aw=np.array(all_walks)

# Plot np_aw and show
plt.plot(np_aw)
plt.show()
# Clear the figure
plt.clf()

# Transpose np_aw: np_aw_t
np_aw_t = np.transpose(np_aw)   ## Now every row in np_all_walks represents the position after 1 throw for the 10 random walks.

# Plot np_aw_t and show
plt.plot(np_aw_t)
plt.show()
```

```python
## Basically, you want to know about the end points of all the random walks you've simulated. 

# Create and plot np_aw_t
np_aw_t = np.transpose(np.array(all_walks))

# Select last row from np_aw_t: ends
ends =np.array(np_aw_t[-1,])

# Plot histogram of ends, display plot
plt.hist(ends)
plt.show()
```
