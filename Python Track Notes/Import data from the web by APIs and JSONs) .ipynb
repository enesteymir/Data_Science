## Loading and exploring a JSON

```python
import json

# Load JSON: json_data
with open("a_movie.json") as ourdata:
    json_data = json.load(ourdata)      ## here, ourdata will be a dictionary

# Print each key-value pair in json_data
for k in json_data.keys():
    print(k + ': ', json_data[k])
```

## What Is API? Import data from APIs

```python
They are everywhere such as Twitter, facebook, Instagram, UBER etc..
An API is a set of protocols and routines for building and interacting with software applications.
API is an acronym and is short for Application Program interface
It is common to pull data from APIs in the JSON file format.
```

```python
to query the API about the movie Hackers, Hugo's query string was 'http://www.omdbapi.com/?t=hackers' and had
a single argument t=hackers.

Note: recently, OMDB has changed their API: you now also have to specify an API key. This means you'll have to
        add another argument to the URL: apikey=72bc447a.
```

```python
# Import requests package
import requests

# Assign URL to variable: url
url = 'http://www.omdbapi.com/?apikey=72bc447a&t=the+social+network'

# Package the request, send the request and catch the response: r
r = requests.get(url)

# Print the text of the response
print(r.text)
```

```python
Your response is actually a JSON, so you can do one step better and decode the JSON. You can then print
the key-value pairs of the resulting dictionary. That's what you're going to do now!
```

```python
# Decode the JSON data into a dictionary: json_data
json_data=r.json()

# Print each key-value pair in json_data
for k in json_data.keys():
    print(k + ': ', json_data[k])
```

### Checking out the Wikipedia API

```python
# Import package
import requests

# Assign URL to variable: url
url='https://en.wikipedia.org/w/api.php?action=query&prop=extracts&format=json&exintro=&titles=pizza'

# Package the request, send the request and catch the response: r
r = requests.get(url)

# Decode the JSON data into a dictionary: json_data
json_data=r.json()

# Print the Wikipedia page extract
# The variable pizza_extract holds the HTML of an extract from Wikipedia's Pizza page as a string
pizza_extract = json_data['query']['pages']['24768']['extract']
print(pizza_extract)
```
