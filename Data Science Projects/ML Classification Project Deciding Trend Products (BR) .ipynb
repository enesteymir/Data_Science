### Choosing Best Model for Classification

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm, tree
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
import xgboost
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn import metrics
```

```python
X=data[['impact_blog', 'impact_forum', 'impact_general',
       'impact_image', 'impact_instagram', 'impact_news', 'impact_review',
       'impact_twitter', 'impact_video', 'twitter retweets', 'instagram likes',
       'sentiment_negative', 'sentiment_neutral', 'sentiment_positive']]

y=data['trend']

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, shuffle=True)
```

```python
# List of Models

classifiers=[]

model1 = svm.SVC()
classifiers.append(model1)

model2 = tree.DecisionTreeClassifier()
classifiers.append(model2)

model3 = RandomForestClassifier()
classifiers.append(model3)

model4 = GradientBoostingClassifier()
classifiers.append(model4)

model5=xgboost.XGBClassifier()
classifiers.append(model5)
```

```python
# Try the models on same data and compare accuracy scores. 
# RandomForest,GradientBoostingClassifier and XGBClassifier can be tried with hypermarameter tunings

for model in classifiers:
    model.fit(X_train, y_train)
    y_pred= model.predict(X_test)
    
    acc = accuracy_score(y_test, y_pred)
    print("Accuracy of %s is %s"%(model, acc))
    
    cm = confusion_matrix(y_test, y_pred)
    print("Confusion Matrix of %s is %s"%(model, cm))
```

```python
# SVC and DecisionTreeClassifier models have lower accuracy. Let's try GradientBoostingClassifier with different paremeters.

# Gradient boosting involves creating and adding trees to the model sequentially.New trees are created to correct the residual
# errors in the predictions from the existing sequence of trees.The effect is that the model can quickly fit, then overfit the
# training dataset.It is prone to overfitting. Try these paremeters and look the training and validation accuracy scores.We can
# see that training score is near to 1 when we put high n_estimators and high max_depth. But still validation score can be max
# nearly 0.77-0.81. So I think it is not enough to take this risk. Lets go on with XGBClassifier and RandomForestClassifier.

# learning_rate (shrinkage factor):This parameter is that new trees are created to correct the residual errors in predictions

lr_list = [0.05, 0.075, 0.1, 0.25, 0.5, 0.75, 1]

for learning_rate in lr_list:
    gb_clf = GradientBoostingClassifier(n_estimators=100,learning_rate=learning_rate,max_features=8,max_depth=5,random_state=0)
    gb_clf.fit(X_train, y_train)

    print("Learning rate: ", learning_rate)
    print("Accuracy score (training): {0:.3f}".format(gb_clf.score(X_train, y_train)))
    print("Accuracy score (validation): {0:.3f}".format(gb_clf.score(X_test, y_test)))
```

### Exploring Random Forest Model Hypermarameters

#### Step 1 :Explore Number of Samples (max_samples)

* The “max_samples” argument can be set to a float between 0 and 1 to control the percentage of the size of the training dataset to make the bootstrap sample used to train each decision tree.
* Setting max_samples to “None” will make the sample size the same size as the training dataset and this is the default.
* A smaller sample size will make trees more different, and a larger sample size will make the trees more similar.
* Standard deviation measures how widely spread data points are.An experiment that yields data with a low standard deviation is said have high precision.The lesser the standard deviation is, the higher the precision is.

```python
from numpy import mean
from numpy import std
from numpy import arange
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold


# def function : get a list of models to evaluate
def get_models():
    models = dict()
    for i in arange(0.1, 1.1, 0.1):  # explore ratios from 10% to 100% in 10% increments
        key = '%.1f' % i
        if i == 1.0:
            i = None
        models[key] = RandomForestClassifier(max_samples=i)
    return models

# def function : evaluate a given model using cross-validation
def evaluate_model(model, X, y):
    cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
    scores = cross_val_score(model, X, y, scoring='accuracy', cv=cv, n_jobs=-1)
    return scores
```

```python
# Get the data (X.y defined before, in feature selection section) and the model

models = get_models()
```

```python
# Choose the best.High accuracy,low std dev. and try it with the real model. By comparing recall,precision and f1-score above.
# You can test this after deciding the other parameters.

results, names = list(), list()
for name, model in models.items():
    scores = evaluate_model(model, X, y)
    results.append(scores)
    names.append(name)
    print('>%s %.3f (%.3f)' % (name, mean(scores), std(scores)))

#Box Plot of Random Forest Bootstrap Sample Size vs. Classification Accuracy
plt.boxplot(results, labels=names, showmeans=True)
plt.show()
```

#### Step 2: Explore Number of Fatures (max_features)

* The number of features that is randomly sampled for each split point is perhaps the most important feature to configure for random forest.

```python
# def function: get a list of models to evaluate,and explore number of features from 1 to 10

def get_models():
    models = dict()
    max_features = [i for i in range(1,10)] + [None]
    for i in max_features:
        models[str(i)] = RandomForestClassifier(max_features=i)
    return models

# def function : evaluate a given model using cross-validation
def evaluate_model(model, X, y):
    cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
    scores = cross_val_score(model, X, y, scoring='accuracy', cv=cv, n_jobs=-1)
    return scores
```

#### Step 3: Explore Number of Trees (n_estimators)

* Typically, the number of trees is increased until the model performance stabilizes. Intuition might suggest that more trees will lead to overfitting, although this is not the case. The default is 100.

```python
# explores the effect of the number of trees with values between 10 to 750. 

# def function: get a list of models to evaluate
def get_models():
    models = dict()
    n_trees = [10, 50, 100,200,300,400,500,600]
    for n in n_trees:
        models[str(n)] = RandomForestClassifier(n_estimators=n)
    return models
 
# def function: evaluate a given model using cross-validation
def evaluate_model(model, X, y):
    cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
    scores = cross_val_score(model, X, y, scoring='accuracy', cv=cv, n_jobs=-1)
    return scores
```

#### Step 4: Explore Tree Depth (max_depth )

* By default, trees are constructed to an arbitrary depth and are not pruned.It is set to None (no maximum depth) by default.

```python
# def function : get a list of models to evaluate
def get_models():
    models = dict()
    # consider tree depths from 1 to 7 and None=full
    depths = [i for i in range(5,25)] + [None]
    for n in depths:
        models[str(n)] = RandomForestClassifier(max_depth=n)
    return models
 
# def function: evaluate a given model using cross-validation
def evaluate_model(model, X, y):
    cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
    scores = cross_val_score(model, X, y, scoring='accuracy', cv=cv, n_jobs=-1)
    return scores
```

```python
import xgboost as xgb
from hyperopt import STATUS_OK, Trials, fmin, hp, tpe
```

* HYPEROPT is a powerful python library that search through an hyperparameter space of values and find the best possible values that yield the minimum of the loss function.

* Bayesian Optimization technique uses Hyperopt to tune the model hyperparameters. Hyperopt is a Python library which is used to tune model hyperparameters.


```python
# try with different n_estimators until reach best score.

space={
        'learning_rate': 0.3,
        'n_estimators': 500, 
        'max_depth': hp.quniform('max_depth', 3, 8, 1),
        'gamma': hp.uniform ('gamma', 0,4),
        'reg_alpha' : hp.quniform('reg_alpha', 0,10,1),
        'reg_lambda' : hp.uniform('reg_lambda', 0,1),
        'colsample_bytree' : hp.uniform('colsample_bytree', 0.5,1),
        'min_child_weight' : hp.quniform('min_child_weight', 0, 4, 1),
      }
```

```python
# defining the objective function

def objective(space):
    
    clf=xgb.XGBClassifier(
                    learning_rate = space['learning_rate'],
                    n_estimators =int(space['n_estimators']), 
                    max_depth = int(space['max_depth']),
                    gamma = space['gamma'],
                    reg_alpha = int(space['reg_alpha']),
                    min_child_weight=int(space['min_child_weight']),
                    colsample_bytree=int(space['colsample_bytree']),
                    reg_lambda = space['reg_lambda'])
                    
    
    evaluation = [( X_train, y_train), ( X_test, y_test)]
    
    clf.fit(X_train, y_train,
            eval_set=evaluation, eval_metric="auc",
            early_stopping_rounds=100,verbose=False)
    

    y_pred = clf.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print ("SCORE:", accuracy)
    return {'loss': -accuracy, 'status': STATUS_OK }
```

```python
# Optimization algorithm

# trials is an object that contains or stores all the relevant information such as hyperparameter,loss-functions for each 
# set of parameters that the model has been trained.

# fmin is an optimization function that minimizes the loss function and takes in 4 inputs - fn, space, algo and max_evals.

# max_evals : indicates that how many times the algorithm tries

trials = Trials()

best_hyperparams = fmin(fn = objective,
                        space = space,
                        algo = tpe.suggest,
                        max_evals = 1000,
                        trials = trials)
```

```python
print("The best hyperparameters are : ","\n")
print(best_hyperparams)
```
